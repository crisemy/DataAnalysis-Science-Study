```python
###Filtering Data Frames. The objective is to filter essentially by rows
stats.head()
```

```python
stats.InternetUsers < 2 # Provides a list of places in which values are meeting that criteria
```

```python
Filter = stats.InternetUsers < 2 # Sets the values in a variable which in this case is a boolean one just because it contains just True/False values
Filter
```

```python
stats[Filter] # Filters out by values that are True. In other words, rows in which the InternetUsers < 2
```

```python
Filter2 = stats.BirthRate > 40
stats[Filter2] ## You can use the following which is the same thing: stats[stats.BirthRate > 40]
```

```python
#Filtering out by 2 columns whereas BirthRate > 40 & InternetUsers < 2
stats[Filter & Filter2] # The other way of doing this is: stats[(stats.BirthRate > 40) & (stats.InternetUsers < 2)]
```

```python
stats[stats.IncomeGroup == "High income" ] # Filters out by IncomeGroups
```

```python
# .iat() --> for intenger location
stats.iat[3, 4] # 3rd row, four column. It does count by column and row
```

```python
# .at() --> For working with labels. It doesn't count
stats.at[2, 'BirthRate']
```
