```python
import numpy as np
```

```python
# Function for creating zero arrays
zerosArray = np.zeros(10)
print(zerosArray) # It prints out float values. We'd need to convert 'm
```

```python
# We'd need to convert 'm into integer values
print(zerosArray.astype(int))
```

```python
# Function for creating ones arrays
onesArray = np.ones(10)
print(onesArray.astype(int)) # It prints out float values. We'd need to convert 'm
```

```python
# Function for creating different array values
preFilledArray = np.full(10,5)
print(preFilledArray.astype(int)) # It prints out float values. We'd need to convert 'm
```

```python
# Using scalar values. It's useful for incrementing values in the elements of the arrays. 
# Note: In plane python this functionality doesn't work
matrix = np.matrix([[1,2,3],[4,5,6]])
print(matrix)
print(matrix + 1)
```

```python
# Using scalar values. Decrement the value of those elements
print(matrix - 3)
```

```python
# Using scalar values. Multipy the value of those elements
print(matrix * 2)
```

```python
# Using scalar values. Divide the value of those elements
print(matrix // 2) # Just one / is for float and // is for integers
```

```python
# Using the transpose option
print(matrix.T)
```

```python
# Usint matmul to multiply values
# For instance: np.matmul(matrix1, matrix2)
```

```python
# Statistic functions. There are tons of those: https://numpy.org/doc/2.2/reference/routines.statistics.html
np.min(matrix)
```

```python
np.max(matrix)
```

```python
np.sum(matrix)
```

```python
np.mean(matrix) # Average of an array
```

```python
np.std(matrix) # Finds the standar diviation of an array
```

```python
np.median(matrix) # Gives the medium value of an array
```
