```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame de prueba
data = {'A単os': [2018, 2019, 2020, 2021, 2022],
        'Ventas': [5000, 7000, 8000, 12000, 15000]}

df = pd.DataFrame(data)

# Graficar los datos
plt.plot(df['A単os'], df['Ventas'], marker='o')
plt.xlabel('A単o')
plt.ylabel('Ventas')
plt.title('Ventas por A単o')
plt.show()
```
