```python
# Adding 1 and 2 arrays
oneDarray = np.array([1,2,3])
print(oneDarray)

print("The {oneDarray} has:" ,oneDarray.ndim, "dimension")

twoDarray = np.array([[1,2,3], [4,5,6]])
print(twoDarray)
print(twoDarray.ndim)
```

```python
# Working with indexes
twoDarray = np.array([[1,2,3],[4,5,6]])
# To index the 1st element of the 2nd array
print(twoDarray[1,0]) # Where 1 is the 1st array and zero is the 1st element of the array. Values for arrays start in Zero
```

```python
threeDarray = np.array([[[1,2,3],[-1,-2,-3]],[[4,5,6],[-4,-5,-6]]])
print(threeDarray[0])
```
