```python
# Univariate Analysis: Analyse the data bases on a variable or column. Helps to search for patterns in the data
# Provides summary statistics only on 1 variable
# There are 2 types of data: 
    # a. Categorical/Discrete. E.g: Spam vs no Spam, male vs female, etc.
    # b. Continous. E.g: Age of population
# Depending upon the 2 types of above data, the EDA is performed differently

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('dataset/iris.csv')
df
```

```python
# Generating the graph
x_axis = df.index
y_axis = df['petal.length']
sns.scatterplot(x=x_axis, y=y_axis, hue=df['variety']) # seaborn library
```

```python
# Displot fuction
sns.displot(df['petal.length'])
```

```python
# Histogram
plt.hist(df['petal.length'])
```

```python
# Categorical Data Analysis
sns.countplot(x=df['variety'])
```

```python
# pie Chart
_labels = ['SETOSA', 'VERSICOLOR', 'VIRGINICA']
plt.pie(df['variety'].value_counts(), labels = _labels)
```
