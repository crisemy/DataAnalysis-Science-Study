```python
# Import it to use it
import yfinance as yf
import pandas as pd
```

```python
# using the download() method for downloading stock data for APPLE
df = yf.download('AAPL', start='2025-12-06', end='2025-12-31')
df.head(10)
```

```python
df.index = pd.to_datetime(df.index)
df
```

```python
# Getting the graph by using Plot option
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.title('Apple stock prices for 5 years')
plt.xlabel('Year')
plt.ylabel('Price')

plt.plot(df['Close'])
```
