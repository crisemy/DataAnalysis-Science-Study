```python
# Boolean list
df = pd.read_csv('dataset/cereals.csv')
# Create a boolean Series with all False values
bool_series = pd.Series([False] * len(df))

# Set the third position to True
bool_series.iloc[2] = True

# Use the boolean Series to select the third row
third_row = df[bool_series]

# Display the third row
print(third_row)
```

```python
# Filtering rows by getting a conditional
# The '&' in Pandas is the and symbol in python. The | one is the or in python
# Take a look at the parenthesis structure which is quite important. Otherwise it'll fail
condition = (df['calories'] > 70) & (df['protein'] == 3) 

# Filter the DataFrame based on the condition
df[condition]
```

```python
# Filtering data using loc()
df.loc[[5,8], 'name':'protein'] # Filters by the 5th and 8th item of the list
```

```python
# Filtering data using iloc()
# iloc() takes row and column positions as arguments and not their labels
df.iloc[0:5, 0:3] # Slice 5 rows and 3 columns 
```

```python
# Another example of using iloc()
df.iloc[[0,2,4], 0:3] # Chooses different rows and the 3 columns
```

```python
# Groupby
import pandas as pd

data = {'Gender': ["female", "male", "female", "male"], 'Score': [85, 88, 95, 80]} # Creating a dictionary
df = pd.DataFrame.from_dict(data)
df
```

```python
# GroupBy option 
df.groupby(df['Gender']).mean() # mean is the average of those female, male scores.
# Other options are sum(), max(), min(), median(), count(), std()(standard deviation)
```
