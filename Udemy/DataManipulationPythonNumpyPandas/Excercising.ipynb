```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a Testing DataFrame
data = {'Years': [2018, 2019, 2020, 2021, 2022],
        'Sales': [5000, 7000, 8000, 12000, 15000]}

df = pd.DataFrame(data)

# Graph your data
plt.plot(df['Years'], df['Sales'], marker='o', linestyle = 'dotted', color = 'r')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Sales per Year')
plt.show()
```
