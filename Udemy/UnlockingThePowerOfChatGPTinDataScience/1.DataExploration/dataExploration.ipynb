```python
# Importing the Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Uploading the Dataset
df = pd.read_csv('housing_data500.csv')
```

```python
# Data Cleaning
# Price: "$2,500,000" → 2500000.0
df['Sale price'] = (
    df['Sale price']
    .replace(r'[\$,]', '', regex=True)   # quita $ y comas
    .astype(float)
)

# Square feets: "3,200" → 3200.0 
df['Square footage'] = (
    df['Square footage']
    .replace(r'[",]', '', regex=True)    # Takes out commas and quotes
    .astype(float)
)
```

```python
# Shows the graph
plt.figure(figsize=(11, 6))
sns.barplot(
    data=df,
    x='Number of bedrooms',
    y='Sales price',
    #estimator='mean',       # puedes cambiar a 'median' si querés menos influencia de mansiones
    estimator='median',
    errorbar='sd',          # barras de error = desviación estándar
    palette='viridis',
    edgecolor='black',
    linewidth=1.2,
    alpha=0.9
)

plt.title('Average Price by Number of Bedrooms', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Name of bedrooms', fontsize=13)
plt.ylabel('Price Size (USD)', fontsize=13)

# Formatting y-axis as currency
formatter = plt.FuncFormatter(lambda x, _: f'${x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)

plt.grid(axis='y', linestyle='--', alpha=0.6)
sns.despine()

plt.tight_layout()
plt.show()
```
